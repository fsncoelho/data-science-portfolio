####  2) Monte um gráfico de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df0[['APGAR1','APGAR5','SEXO','GESTACAO','ESTCIVMAE']].head()
```

```python
plt.rc('figure', figsize=(9, 10))
# fig, axes = plt.subplots(3,2)
fig = plt.figure(layout='constrained')
axes = fig.subplot_mosaic(
    """
    AB
    CD
    EE
    """
)

sns.countplot(ax = axes['A'], data=df0, x='APGAR1')
sns.countplot(ax = axes['B'], data=df0, x='APGAR5')
sns.countplot(ax = axes['C'], data=df0, x='SEXO')
sns.countplot(ax = axes['D'], data=df0, x='ESTCIVMAE')
sns.countplot(ax = axes['E'], data=df0, x='GESTACAO')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(data=df0, x='CONSULTAS', y='APGAR5')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig = plt.figure(layout='constrained')
axes = fig.subplot_mosaic(
    """
    AB
    CC
    DD
    """
)

sns.countplot(ax = axes['A'], data=df0, x='SEXO')
sns.countplot(ax = axes['C'], data=df0, x='ESTCIVMAE')
sns.countplot(ax = axes['B'], data=df0, x='CONSULTAS')
sns.countplot(ax = axes['D'], data=df0, x='GESTACAO')

# axes['C'].set_xticklabels(['1:Solteira', '2:Casada', '3:Viúva', '4:Divorciada', [], '9:Ignorado'])
```
