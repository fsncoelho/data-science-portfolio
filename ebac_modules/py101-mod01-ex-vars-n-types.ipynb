## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
cancao_low = cancao.lower() # save str into a new lower case str 
print(cancao_low)

n = cancao_low.count('roda') # count a specif str that repeats
print(n)

idx = cancao_low.find('roda') # find fisrt occurance for 'roda'
print(idx)

cancao.encode() # returns encoded version of a str
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

# get selic position in str
x = '2,75%'
idx1 = noticia.find(x)
print(idx1)

# extract selic as str
selic_str = noticia[idx1:idx1+5]
print(selic_str)

# replace '%' and change decimal separator ',' and save selic in a float var
selic_float = float(selic_str.replace('%','').replace(',','.')) / 100
print(selic_float)

# get ano position in str
x2 = '6'
idx2 = noticia.find(x2)
print(idx2)

# extract ano as str
ano_str = noticia[idx2]
print(ano_str)
print(type(ano_str))

# covert to integer
ano_int = int(ano_str)
print(ano_int)
print(type(ano_int))

print(f'O valor da selic é {selic_float}, isto é {selic_str}, num período de {ano_str} anos.')
```
