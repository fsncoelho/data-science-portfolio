```python
# boxplot com seaborn
sns.boxplot(data=df0,y='PESO')
```

```python
# plt.rc('figure',figsize=(10,5))
sns.histplot(
    data=df0
    ,x='PESO'
    ,bins=100
    ,alpha=0.3
    ,element='step'
    ,kde=True
)
```
